```
pid = train_df["PassengerId"].tolist()
sur = train_df["Survived"].tolist()
pcls = train_df["Pclass"].tolist()
age = train_df["Age"].fillna(0).tolist()
sib = train_df["SibSp"].tolist()
parch = train_df["Parch"].tolist()
fare = train_df["Fare"].tolist()

pid_list = ["PassengerId", len(pid), stat.mean(pid), stat.pstdev(pid), min(pid), np.percentile(pid, 25), np.percentile(pid, 50), np.percentile(pid, 75), max(pid)]
sur_list = ["Survived", len(sur), stat.mean(sur), stat.pstdev(sur), min(sur), np.percentile(sur, 25), np.percentile(sur, 50), np.percentile(sur, 75), max(sur)]
pcls_list = ["Pclass", len(pcls), stat.mean(pcls), stat.pstdev(pcls), min(pcls), np.percentile(pcls, 25), np.percentile(pcls, 50), np.percentile(pcls, 75), max(pcls)]
age_list = ["Age", len(age), stat.mean(age), stat.pstdev(age), min(age), np.percentile(age, 25), np.percentile(age, 50), np.percentile(age, 75), max(age)]
sib_list = ["SibSp", len(sib), stat.mean(sib), stat.pstdev(sib), min(sib), np.percentile(sib, 25), np.percentile(sib, 50), np.percentile(sib, 75), max(sib)]
parch_list = ["Parch", len(parch), stat.mean(parch), stat.pstdev(parch), min(parch), np.percentile(parch, 25), np.percentile(parch, 50), np.percentile(parch, 75), max(parch)]
fare_list = ["Fare", len(fare), stat.mean(fare), stat.pstdev(fare), min(fare), np.percentile(fare, 25), np.percentile(fare, 50), np.percentile(fare, 75), max(fare)]

print(tabulate([pid_list, sur_list, pcls_list, age_list, sib_list, parch_list, fare_list],
               headers=["Feature", "Length", "Mean", "Std", "Min", "25%", "50%", "75%", "Max"], tablefmt='orgtbl'))
```

```
age = train_df["Age"].fillna(-1).tolist()
sur = train_df["Survived"].tolist()

d = []
d_d = []

for x in range(0, len(age)):
  if age[x] == -1:
    continue
  if sur[x] == 1:
    d.append(age[x])
  else:
    d_d.append(age[x])

bins = np.arange(0, 100, 5)
plt.xlim([0, max(d)+5])
plt.hist(d, bins=bins)
plt.title("Survival = 1")
plt.xlabel("Age")
plt.ylabel("Count")
plt.yticks(np.arange(0, 100, step=20))
```
