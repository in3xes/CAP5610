**Task 2**

Map from [x1, x2] to [x1, x1x2]

[−1,−1]  -> [−1, +1] (negative)

[−1, +1] -> [−1,−1]  (positive)

[+1,−1]  -> [+1,−1]  (positive)

[+1, +1] -> [+1, +1] (negative)

```
data = np.array([
    [-1, 1],
    [1, 1],
])
x, y = data.T
plt.scatter(x,y, color='red')

for xs,ys in zip(x,y):
    label = f"({xs},{ys})(positive)"
    plt.annotate(label,
                 (xs,ys),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')


data = np.array([
    [-1, -1],
    [1, -1],
])
x, y = data.T
plt.scatter(x,y, color='blue')
for xs,ys in zip(x,y):
    label = f"({xs},{ys})(negative)"
    plt.annotate(label,
                 (xs,ys),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

x = np.linspace(-2,2,100)
y = x*0
plt.plot(x, y, 'grey')

y = np.linspace(-2,2,100)
x = y*0
plt.plot(x, y, 'grey')
plt.show()
```

Based on the above plot we can see that y = 0(x1x2 = 0) is the maximum margin separtor. It has a margin of 1.

```
data = np.array([
    [1, 1],
    [2, 2],
    [2, 0],    
])
x, y = data.T
plt.scatter(x,y, color='red')

for xs,ys in zip(x,y):
    label = f"({xs},{ys})(+)"
    plt.annotate(label,
                 (xs,ys),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')


data = np.array([
    [0, 0],
    [1, 0],
    [0, 1],    
])
x, y = data.T
plt.scatter(x,y, color='blue')
for xs,ys in zip(x,y):
    label = f"({xs},{ys})(-)"
    plt.annotate(label,
                 (xs,ys),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

x = np.linspace(-1,3,100)
y = x*0
plt.plot(x, y, 'grey', linewidth=0.5)

y = np.linspace(-1,3,100)
x = y*0
plt.plot(x, y, 'grey', linewidth=0.5)



x = np.linspace(-1,3,100)
y = x * -1 + 1
plt.plot(x, y, 'blue', linewidth=0.5)

x = np.linspace(-1,3,100)
y = x * -1 + 2
plt.plot(x, y, 'red', linewidth=0.5)

x = np.linspace(-1,3,100)
y = x * -1 + 1.5
plt.plot(x, y, 'black')


plt.show()
```

We can see that the line touching the (-) points has a slope of -1 and the equation is `y = x * -1 + 1`

The line that is touching the red points is `y = x * -1 + 2`

The maximum margin hyperplane(line) has the same slope. The equation is `y = x * -1 + 1.5`

New feature vectors based on [1, sqrt(2)x, x^2 ] are:

(1, 0, 0), 

(1,−sqrt(2), 1), 

(1,sqrt(2), 1)

```
from mpl_toolkits.mplot3d import Axes3D
from math import sqrt


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

root2 = sqrt(2)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


for xs, ys, zs in [(1, root2, 1), (1, -1 * root2, 1)]:
    ax.scatter(xs, ys, zs, color='blue')

for xs, ys, zs in [(1, 0, 0)]:
    ax.scatter(xs, ys, zs, color='red')


plt.show()
```

We can see that (-) negative are in z=1 plane and (+) point is z=0 place.

```
from mpl_toolkits.mplot3d import Axes3D
from math import sqrt

root2 = sqrt(2)

def f(x, y):
    return x*0 + y*0 + 0.5
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title('wireframe')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')


for xs, ys, zs in [(1, root2, 1), (1, -1 * root2, 1)]:
    ax.scatter(xs, ys, zs, color='blue')

for xs, ys, zs in [(1, 0, 0)]:
    ax.scatter(xs, ys, zs, color='red')


plt.show()
```

```
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

test = pd.read_csv('/content/test.csv')
train = pd.read_csv('/content/train.csv')

combined = [train, test]

#preprocessing:

#From homework 1 & 2

from numpy.random import uniform

# Fill null in Fare to median
for df in combined:
  median_fare = df['Fare'].median()
  df['Fare'] = df['Fare'].fillna(median_fare)

for t in combined:
  train_fare = t["Fare"].tolist()

  for x in range(0, len(train_fare)):
    if t.at[t.index[x], 'Fare'] > -0.001 and t.at[t.index[x], 'Fare'] <= 7.91:
      t.at[t.index[x], 'Fare'] =  0
    elif t.at[t.index[x], 'Fare'] > 7.91 and t.at[t.index[x], 'Fare'] <= 14.454:
      t.at[t.index[x], 'Fare'] =  1
    elif t.at[t.index[x], 'Fare'] > 14.454 and t.at[t.index[x], 'Fare'] <= 31.0:
      t.at[t.index[x], 'Fare'] =  2
    elif t.at[t.index[x], 'Fare'] > 31.0 and t.at[t.index[x], 'Fare'] <= 512.329:
      t.at[t.index[x], 'Fare'] =  3  

# Fill null in embarked and make them numerical
embarked_to_num_map = {'S': 0, 'C': 1, 'Q': 2}
for df in combined:
  df['Embarked'] = df['Embarked'].fillna('S')
  df['Embarked'] = df['Embarked'].map(embarked_to_num_map)


# Change sex to numerical
sex_to_num_map = {'female': 1, 'male': 0}
for df in combined:
  df['Sex'] = df['Sex'].map(sex_to_num_map).astype(int)

# Adgust null in Age based on median and std
for df in combined:
  std_age = df['Age'].std()
  median_age = df['Age'].median() 
  df['Age'] = df['Age'].fillna(uniform(std_age, median_age))
  df.loc[ df['Age'] <= 15, 'Age'] = 0
  df.loc[(df['Age'] > 15) & (df['Age'] <= 30), 'Age'] = 1
  df.loc[(df['Age'] > 30) & (df['Age'] <= 45), 'Age'] = 2
  df.loc[(df['Age'] > 45) & (df['Age'] <= 60), 'Age'] = 3
  df.loc[ df['Age'] > 60, 'Age'] = 4  




# feature selection

drop_elements = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp']
train = train.drop(drop_elements, axis = 1)
test  = test.drop(drop_elements, axis = 1)

label = train["Survived"]
train_d = train.drop("Survived", axis=1)
# train_data,test_data,train_labels,test_labels=train_test_split(train_d, label, test_size=0.3, random_state=42)

# print(train_data.shape, test_data.shape, train_labels.shape, test_labels.shape)

train_data = train_d
train_labels = label
```

```
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, GridSearchCV, cross_validate



svm_param = {'kernel': ['rbf'], 'C': [10,20,30,40,50,60,70], 'gamma': [0.1, 0.01, 0.001, 0.001], 'probability': [True]}


svm = GridSearchCV(SVC(), svm_param, cv=5)
svm.fit(train_data, train_labels)
print("Best accuracy: ", svm.best_score_)
print("Params: ", svm.best_params_)

```
